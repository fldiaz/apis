```python
import ssl
import sys
from SPARQLWrapper import SPARQLWrapper, JSON

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
def get_results(endpoint_url, query):
    user_agent = "WDQS-example Python/%s.%s" % (sys.version_info[0], sys.version_info[1])
    sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    return sparql.query().convert()
def nulls(df):
    nulls = pd.DataFrame(df.isnull().sum().sort_values(), columns=['nulls'])
    nulls['porcentaje'] = round(100*nulls['nulls'] / len(df), 2)
    nulls.sort_values('porcentaje')
    return nulls
```

```python
def datos_wiki(q):
    endpoint_url = "https://query.wikidata.org/sparql"
    resultados = []
    query=('SELECT distinct  ?person ?personLabel ?dod  ?countryLabel ?twitterName'\
    + '(GROUP_CONCAT(DISTINCT ?occupationLabel; separator = ", ") as ?occupationLabels)' \
    +'(GROUP_CONCAT(DISTINCT ?educatedLabel; separator = ", ") as ?educatedLabels )' \
    +'(GROUP_CONCAT(DISTINCT ?nomineeLabel; separator = ", ") as ?nomineeLabels)' \
    +'(GROUP_CONCAT(DISTINCT ?awardLabel; separator = ", ") as ?awardLabels)' \
    +'WHERE { ?person rdfs:label' + '"{}"'.format(q) +  '@en .' \
    +' ?person rdfs:label ?personLabel  filter (lang(?personLabel) = "en").' \
    +'OPTIONAL {?person wdt:P570 ?dod } .' \
    +'OPTIONAL {?person wdt:P2002 ?twitterName}.' \
    +'OPTIONAL {?person wdt:P27 ?country .?country rdfs:label ?countryLabel filter (lang(?countryLabel) = "en")}.' \
    +'OPTIONAL {?person wdt:P106 ?occupation .?occupation rdfs:label ?occupationLabel filter (lang(?occupationLabel) = "en") } .' \
    +'OPTIONAL {?person wdt:P1411 ?nominee .?nominee rdfs:label ?nomineeLabel  filter (lang(?nomineeLabel) = "en") }.' \
    +'OPTIONAL {?person wdt:P69 ?educated .?educated rdfs:label ?educatedLabel filter (lang(?educatedLabel) = "en") }.' \
    +'OPTIONAL {?person wdt:P166 ?award .?award rdfs:label ?awardLabel   filter (lang(?awardLabel) = "en")}} GROUP BY ?person ?personLabel ?dod  ?countryLabel ?twitterName') 
    
    data = get_results(endpoint_url, query)
    for item in data['results']['bindings']:
        resultados.append(item)
        #print('Agregando datos de wikipedia : {}'.format())
    return resultados
```

```python
#+'SERVICE wikibase:label{ bd:serviceParam wikibase:language "[AUTO_LANGUAGE]"}
datos = []
for autor in list:
    print(autor)
    data = datos_wiki(autor)
    #df=pd.json_normalize(data)
    datos.append(data)
```

```python
outdf = pd.DataFrame()
for i in range(len(datos)):
    for e in range (len(datos[i])):
        row = pd.json_normalize(datos[i][e])
        outdf = pd.concat([outdf, row])
```

```python
outdf.to_csv('Datos_wiki_people.csv')
```
