```python
import requests
import pandas as pd

def query_raw(text):
    url="https://bern.korea.ac.kr/plain"
    ner=requests.post(url, data={'sample_text': text}, verify=False).json()
    df=pd.json_normalize(ner)
    outdf = pd.DataFrame()
    for i in ner['denotations']:
        row = pd.json_normalize(i)
        outdf = pd.concat([outdf, row])
        outdf.reset_index(drop=True, inplace=True)

    
    return outdf

#if __name__ == '__main__':
 #   print(query_raw("YOUR TEXT HERE"))
```
